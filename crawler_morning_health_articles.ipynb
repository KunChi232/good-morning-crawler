```python
import requests
from bs4 import BeautifulSoup
import csv

url = "https://www.edh.tw/tag/%E4%BD%9C%E6%81%AF/1"
base_url = "https://www.edh.tw"

r = requests.get(url)
soup = BeautifulSoup(r.text, "html.parser")

output_file = open('作息_早安健康.csv', 'w', newline='', encoding="utf_8_sig")

# 建立 CSV 檔寫入器
writer = csv.writer(output_file)

# 寫入一列資料
writer.writerow(['文章標題', '摘要', '文章url', '縮圖url', 'paragraph'])

def _slice(content, length = 200):
    stop = ['。', '!', '！', "?", "？"]
    para = list()
    start_idx = 0
    while start_idx < len(content) - 1:
        end_idx = start_idx + length
        if(end_idx > len(content) - 1):
            end_idx = len(content) - 1
        last_word = content[end_idx]
        while last_word not in stop and (end_idx+1) < len(content):
            last_word = content[end_idx + 1]
            end_idx = end_idx + 1
        para.append(content[start_idx : end_idx + 1])
        start_idx = end_idx + 1
    return para

def write_csv(title, abstract, url, thumbnail_url, paragraph):
    # 寫入列資料
    writer.writerow([title, abstract, url, thumbnail_url, paragraph])
    
    # 輸出
    print("url: " + base_url + s["href"] + ", title: " + s.text + ", abstract: " + str(abstract_sel[index].text).strip() + ", img: " + str(thumbnail_sel[index]["src"]))
    print("--------------------")

while url != None:
    print("=====================================================")
    print("這一頁的url: " + url)

    main = soup.find('main')
    
    # get title
    title_sel = main.select(".title a")

    # get abstract
    abstract_sel = main.select(".list-article a.content")
    
    # get thumbnail
    thumbnail_sel = main.select(".list-img img")
    
    # check array size
    print("abstract array size: " + str(len(abstract_sel)) + ", title size: " + str(len(title_sel)) + ", thumbnail size: " + str(len(thumbnail_sel)))
    
    # get page info
    for index, s in enumerate(title_sel):
        next_url = base_url + s['href']
        contents = list()
        while next_url != None:
            r_in = requests.get(next_url)
            soup_in = BeautifulSoup(r_in.text, "html.parser")
            main_in = soup_in.find('main')
            if(main_in == None):
                break
            content = main_in.select('#article_page')
            all_text = content[0].findAll(text=True)
            all_text = [item.strip() for item in all_text if str(item)]
            contents += all_text
            try:
                next_url = main_in.find('a', {'rel':'next'})['href']
            except :
                next_url = None
#         print(''.join(contents))
        write_csv(s.text, str(abstract_sel[index].text).strip(), (base_url + s["href"]), str(thumbnail_sel[index]["src"]), _slice(''.join(contents)))

    # next page
    try:
        url = soup.find("link",{"rel":"next"})['href']
        r = requests.get(url)
        soup = BeautifulSoup(r.text, "html.parser")
    except:
        url = None

output_file.close()
print("===== 爬蟲結束 =====")
```
